## Population

```python
def plot_measure(df, measure_idx, x_label_shift, y_label_shift, title, year_idx_range=slice(1, -1)):
    years = df.iloc[measure_idx, year_idx_range].index
    if year_idx_range.stop is None or year_idx_range.stop == len(df.columns):
        years = [str(y) for y in years]
    measure_values = df.iloc[measure_idx, year_idx_range].values
    
    plt.figure(figsize=(10, 5))
    plt.plot(years, measure_values, marker='o')
    plt.xticks(list(years))
    plt.title(title)
    
    if year_idx_range.stop is not None and year_idx_range.stop != len(df.columns):
        for ye, val in zip(years, measure_values):
            plt.text(ye+x_label_shift, val+y_label_shift, '{:.2f}'.format(val))
            
    plt.show()
    

plot_measure(GG_city_df, 0, -2, 2000, "Garden Grove's population from 1970 to 2010")
```

```python
pop_density = pd.concat([pd.Series(data=['Population density'], index=['Measure']), 
                         GG_city_df.iloc[0, 1:]/17.98])
GG_city_df = GG_city_df.append(pop_density, ignore_index=True)
```

```python
plot_measure(GG_city_df, 25, -2, 55, "Garden Grove's density from 1970 to 2010")
```

```python
def plot_mul_measures(df1, measure_idx_list1, df2=None, measure_idx_list2=[], 
                      year_idx_range=slice(1, -1), title=None, figsize=(10, 5)):
    plt.figure(figsize=figsize)
    legends = []
    
    years = [str(y) for y in df1.iloc[measure_idx_list1[0], year_idx_range].dropna().index]
    
    for measure_idx1 in measure_idx_list1:
        measure_values = df1.iloc[measure_idx1, year_idx_range].dropna().values
        measure_name = df1.iloc[measure_idx1, 0]
        legends.append(measure_name)
        plt.plot(years, measure_values, marker='o')
        
    for measure_idx2 in measure_idx_list2:
        measure_values = df2.iloc[measure_idx2, year_idx_range].dropna().values
        measure_name = df2.iloc[measure_idx2, 0]
        legends.append(measure_name)
        plt.plot(years, measure_values, marker='o')
        
    plt.xticks(list(years))
    plt.legend(legends)
    plt.title(title)
```

```python
plot_mul_measures(GG_city_df, [26], Orange_df, [27], 
                  title='Density percent change of Garden Grove and Orange county from 1980 to 2010')
plt.legend(['Garden Grove', 'Orange county'])
plt.show()
```

```python
from matplotlib import colors as plt_colors

def plot_stacked_bar(df, measure_idx_range, year_idx_range=slice(1,-1), 
                     spec_measure_idx='all', figsize=(9.2, 5)):
    years = [str(y) for y in df.columns[year_idx_range]]
    
    data = df.iloc[measure_idx_range, year_idx_range].fillna(0)
    data_cum = data.cumsum(axis=0)
    
    measure_names = GG_city_df.iloc[measure_idx_range, 0]
    
    if len(measure_names) == 5:
        measure_colors = [plt_colors.to_rgb('steelblue'), plt_colors.to_rgb('peru'),
                          plt_colors.to_rgb('mediumseagreen'), plt_colors.to_rgb('lemonchiffon'),
                          plt_colors.to_rgb('indianred')]
    else:
        measure_colors = plt.rcParams['axes.prop_cycle'].by_key()['color']

    fig, ax = plt.subplots(figsize=figsize)
    ax.invert_yaxis()
    ax.xaxis.set_visible(False)
    ax.set_xlim(0, np.sum(data, axis=0).max())
        
    for i, (measure, color) in enumerate(zip(measure_names, measure_colors)):
        widths = data.iloc[i, :]
        starts = data_cum.iloc[i, :] - widths
        ax.barh(years, widths, left=starts, height=0.5,
                label=measure, color=color)
        if spec_measure_idx == 'all' or i in spec_measure_idx:
            xcenters = starts + widths / 2
        
            r, g, b = color
            text_color = 'white' if r * g * b < 0.5 else 'darkgrey'     
            for y, (x, c) in enumerate(zip(xcenters, widths)):
                try:
                    ax.text(x, y, str(int(c)), ha='center', va='center', color=text_color)
                except ValueError:
                    pass
        
    
    ax.legend(ncol=len(measure_names)//2, bbox_to_anchor=(-0.01, 1),
              loc='lower left', fontsize=11)
    
    return fig, ax

plot_stacked_bar(GG_city_df, slice(28, 33), spec_measure_idx=[0, 4]) 
plt.title("Garden Grove - Racial groups' population percentages", y=1.23)
plt.show()
```

```python
def plot_grouped_bar(df1, measure_idx1, df2, measure_idx2, labels, year_idx_range=slice(1, -1), title=None, figsize=(4.5, 3)):
    data1 = df1.iloc[measure_idx1, year_idx_range].dropna()
    data2 = df2.iloc[measure_idx2, year_idx_range].dropna()
    
    years = [str(y) for y in data1.index]
    
    ind = np.arange(len(data1))  # the x locations for the groups
    width = 0.35  # the width of the bars

    fig, ax = plt.subplots(figsize=figsize)
    rects1 = ax.bar(ind - width/2, data1, width, label=labels[0])
    rects2 = ax.bar(ind + width/2, data2, width, label=labels[1])

    # Add some text for labels, title and custom x-axis tick labels, etc.
    ax.set_title(title)
    ax.set_xticks(ind)
    ax.set_xticklabels(years)
    ax.legend()

    plt.show()
    
plot_grouped_bar(GG_city_df, 33, Orange_df, 28, ['Garden Grove', 'Orange county'], year_idx_range=slice(1, None),
                 title='Percentage of households with income from $25k')

plot_grouped_bar(GG_city_df, 34, Orange_df, 29, ['Garden Grove', 'Orange county'], year_idx_range=slice(1, None),
                 title='Poverty percentage')
```
