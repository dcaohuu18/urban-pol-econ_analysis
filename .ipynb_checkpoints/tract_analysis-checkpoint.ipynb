## White and Asian percentage

```python
exp = grouped_tracts.get_group('Census Tract 761.03')
asian_pops = exp.iloc[4:11, 2:-1].sum().values
asian_percentages = 100*asian_pops/exp.iloc[0, 2:-1]
asian_percentages.index = (['Asian and PI percentage {}'.format(i) for i in range(1970, 2020, 10)])
asian_percentages = pd.concat([pd.Series(data=['Census Tract 761.03'], index=['Tract_name']), asian_percentages])
asian_percentages
```

```python
selected_df = pd.DataFrame()
for tract_name, tract in grouped_tracts:
    tract = tract.reset_index(drop=True)
    tract_asian_pops = tract.iloc[4:11, 2:-1].sum().values
    tract_asian_percentages = 100*tract_asian_pops/tract.iloc[0, 2:-1]
    tract_asian_percentages.index = (['Asian and PI percentage {}'.format(i) for i in range(1970, 2020, 10)])
    tract_asian_percentages = pd.concat([pd.Series(data=[tract_name], index=['Tract_name']), 
                                         tract_asian_percentages])
    selected_df = selected_df.append(tract_asian_percentages, ignore_index=True)
```

```python
def get_percentages(sub_measure_idx, new_measure_name, tract_group, tot_measure_idx=0):
    percentages_df = pd.DataFrame()
    for tract_name, tract in tract_group:
        tract = tract.reset_index(drop=True)
        tract_percentages = 100*tract.iloc[sub_measure_idx, 2:]/tract.iloc[tot_measure_idx, 2:]
        tract_percentages.index = (['{} {}'.format(new_measure_name, y) for y in range(1970, 2020, 10)]+
                                   ['{} 2008-2012'.format(new_measure_name)])
        percentages_df = percentages_df.append(tract_percentages, ignore_index=True)
    percentages_df.dropna(axis=1, how='all', inplace=True)
    return percentages_df
```

```python
white_percentages_df = get_percentages(1, 'White percentage', grouped_tracts)
selected_df = pd.concat([selected_df, white_percentages_df], axis=1)
```

# Regression

## Regression between race percentage and poverty

```python
from sklearn import linear_model

def run_regression(df, x_col, y_col):
    df = df.dropna(subset=[x_col, y_col], axis=0)
        
    regr = linear_model.LinearRegression()
    regr.fit(df[x_col].values.reshape(-1, 1), df[y_col].values)
    
    return regr.predict(df[x_col].values.reshape(-1, 1))
```

```python
white_percent = selected_df['White percentage 1990']
poverty_percent = selected_df['Poverty percentage 1990']
predicted_pov = run_regression(selected_df, 'White percentage 1990', 'Poverty percentage 1990')
```

```python
plt.scatter(white_percent, poverty_percent)
plt.plot(white_percent.dropna(), predicted_pov)
```

```python
def plot_multi_reg(df, measure_dict, title=None, figsize=(9.2, 5)):
    legends =[]
    plt.figure(figsize=figsize)
    for x_col, y_col in measure_dict.items():
        X = df[x_col]
        predicted_y = run_regression(df, x_col, y_col)
        plt.plot(X.dropna(), predicted_y)
        legends.append(x_col.split()[-1])
    plt.legend(legends)
    plt.xlabel(' '.join(x_col.split()[:-1]))
    plt.ylabel(' '.join(y_col.split()[:-1]))
    plt.title(title)
```

```python
white_pov_dict = {
    'White percentage 1980': 'Poverty percentage 1980',
    'White percentage 1990': 'Poverty percentage 1990',
    'White percentage 2000': 'Poverty percentage 2000',
    'White percentage 2010': 'Poverty percentage 2008-2012'
}

plot_multi_reg(selected_df, white_pov_dict, title='Regression between white percentage and poverty')
```

```python
from scipy import stats 

def get_corr_coef(df, measure_dict):
    for x_col, y_col in measure_dict.items():
        pearson_coef, p_value = stats.pearsonr(df[x_col].dropna(), df[y_col].dropna())
        print("Correlation Coefficient between {} and {}: {:.2f}".format(x_col, y_col, pearson_coef), 
              "and a P-value of: {:.2f}".format(p_value))
        
get_corr_coef(selected_df, white_pov_dict)
```

```python
asian_pov_dict = {
    'Asian and PI percentage 1980': 'Poverty percentage 1980',
    'Asian and PI percentage 1990': 'Poverty percentage 1990',
    'Asian and PI percentage 2000': 'Poverty percentage 2000',
    'Asian and PI percentage 2010': 'Poverty percentage 2008-2012'
}

plot_multi_reg(selected_df, asian_pov_dict, title='Regression between Asian and Pacific Islander percentage and poverty')
```

```python
get_corr_coef(selected_df, asian_pov_dict)
```

## Regression between race percentage and unemployment rate

```python
asian_unemp_dict = {
    'Asian and PI percentage 1980': 'Unemployment percentage 1980',
    'Asian and PI percentage 1990': 'Unemployment percentage 1990',
    'Asian and PI percentage 2000': 'Unemployment percentage 2000',
    'Asian and PI percentage 2010': 'Unemployment percentage 2008-2012'
}

plot_multi_reg(selected_df, asian_unemp_dict, 
               title='Regression between Asian and Pacific Islander percentage and Unemployment percentage')
```

```python
get_corr_coef(selected_df, asian_unemp_dict)
```

**Comment:** A major regression around 2010: effect of the Great Recession in 2008. The crisis hit everyone hard but it was most challenging for neighborhoods with a high percentage of Asian Americans. However, there was not a disproportionate effect --> the slope in 2010 is similar to that in 2000. However, this statement should be treated with caution due to the low correlation coefficient.

```python
asian_percent = selected_df['Asian and PI percentage 2010']
unemp_percent = selected_df['Unemployment percentage 2008-2012']
predicted_unemp = run_regression(selected_df, 'Asian and PI percentage 2010', 'Unemployment percentage 2008-2012')
plt.figure(figsize=(9.2, 5))
plt.scatter(asian_percent, unemp_percent)
plt.plot(asian_percent.dropna(), predicted_unemp, color='indianred')
plt.xlabel('Asian and PI percentage')
plt.ylabel('Unemployment percentage')
plt.title('Relationship between Asian and Pacific Islander percentage and Unemployment rate in 2010')
plt.show()
```

**Comment:** Integrated neighborhoods seems to be the most affected areas. Predominantly asian neihborhoods were less affected than expected --> self-sufficient communities? Predominantly white were the least affected.
